```python
##Libraries needs to run the script##
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import pandas as pd
from scipy import interpolate
import math
```

```python
## read the input files using pandas##
df = pd.read_csv('../curvature_lipids.dat', delim_whitespace=True, names= ["Lipid", "Value", "SE"] )
```

```python
fig = plt.figure(1034, figsize=(14,10))
ax = fig.add_subplot(111)
Lipid = df['Lipid']
STD = df['SE']

#Value = df['Value']
# Read value for 2 decimal point
Value = round(df['Value'], 3)

# This is the location for the annotated text
i = 0.20
j = 0.1
# Annotating the bar plot with the values in column(Value)
for i in range(len(Lipid)):
    #plt.annotate(Value[i], (-0.1 + i, Value[i] + j))
    plt.annotate(Value[i], (i-0.15, Value[i]-0.35),rotation=90, color='green', fontsize=50, fontname='arial')
#thick axes
from pylab import *
rc('axes', linewidth=3)    
plt.xticks(rotation=90, horizontalalignment="center", fontsize=50, fontname='arial')
plt.ylim([-0.7, 0.3501])
plt.yticks(np.arange(-0.7, 0.3501, 0.1 ), fontsize=50, fontname='arial')
#plt.title("Membrane bending moduli", fontsize=24, fontname='arial')
plt.ylabel(r"C$_0$ [nm$^{-1}$] ReSIS/GLD",loc='center',fontsize=50, fontname='arial',labelpad=15)
plt.tick_params(axis ='both', which= 'major', labelsize=50, direction='in', length=18, top=1, right=1, width=1.7, pad=15)
plt.bar(Lipid, Value, yerr=STD, width= 0.30, align='center',color='green', edgecolor = 'black', capsize=10)
# Saving the plot as a 'png'
#plt.savefig('Barplot_new.pdf',bbox_inches='tight', dpi=600)
plt.title("(B)", loc = 'left', y=1.07, x=-0.04, fontsize=50, fontname='arial')
plt.savefig('Figure5B.png', bbox_inches='tight', dpi=600)
# Displaying the bar plot
plt.show()
```
